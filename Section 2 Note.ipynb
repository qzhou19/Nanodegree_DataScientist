### Upload Package to PyPi

__Prerequisites__  
A folder named with the package name, assume X_Y:  
- general.py # define Dist class - actual pacakage code
- gaussian.py  # define Dist inherited Gaussian class - actual pacakage code  
- binomial.py  # define Dist inherited Binomial class - actual pacakage code
- $__init__.py$
- license.txt # copyright info, opensource.org mit  
- README.md # summary of packages and explain of files, how the package are   
- __setup.cfg__ # shown below  

Each py file should be able to run itself and include all libraries needed

$__init__.py$ file (as in above example)  
from .gaussian import Gaussian  
from .binomial import Binomial  

__setup.cfg__ contents  
[metadata]  
description-file=README.md



A __setup.py__ in the same level as the folder    
from setuptools import setup  

setup(name='X_Y', version='0.1', description='package description', packages=['X_Y'], author='xx', zip_safe=False)
      
### Uploading

In terminal, type following  
_python setup.py sdist_ # this will generate packages to be uploaded  
_pip install twine_  

#### Uploading to pypi test repoistory

#### routes.py file codes:
from myappfolder import app  
import json, plotly  
from flask import render_template  
from data_working_scripts.data_working import return_figures   

@app.route('/')  
@app.route('/index')  
def index():  
figures = return_figures()
ids = ['figure-{}'.format(i) for i, _ in enumerate(figures)] # plot ids for the html id tag

figuresJSON = json.dumps(figures, cls=plotly.utils.PlotlyJSONEncoder) # Convert the plotly figures to JSON for javascript in html template
return render_template('index.html', ids=ids, figuresJSON=figuresJSON)

#### html file 
__head__ section need to import script files needed from plotly and bootstrp  
__footer__ section need to include the script to call the visuals
<script type="text/javascript">
    // plots the figure with id
    // id must match the div id above in the html
    var figures = {{figuresJSON | safe}};
    var ids = {{ids | safe}};
    for(var i in figures) {
        Plotly.plot(ids[i],
            figures[i].data,
            figures[i].layout || {});
    }
</script>

#### Deployment steps
* in above dir terminal, run myapp.py to ensure it can run locally  
* update python
* create virtual environment 
* pip install necessary packages for the environment: flask, pandas, ploty, gunicorn  
* intall heroku command line tools/check installation by 'heroku --verison'
* login heroku 'heroku login'
* remove app.run() line of code from myapp.py
* go to above dir create Procfile 'touch Procfile', put in following 'web gunicorn myapp:app'
* create requirements file 'pip freeze > requirements.txt'
* setup git: 'git init; git add .; git config --global user.email x@y.com; git config --global user.name "x"'
* commit the files 'git commit -m "first commit"'
* create heroku app: 'heroku create my-app-name'
* check git and upload: 'git remote -v; git push heroku master'
* check the link for the webpage, something like https://app-name.heroku.com
